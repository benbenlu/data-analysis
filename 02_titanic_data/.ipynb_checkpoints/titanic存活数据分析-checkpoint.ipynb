#### 生存列和其他列的相关性
用numpy的corrcoef函数得出相关系数矩阵，然后再计算矩阵的行列式
或者自己也可编写计算皮尔逊积矩相关系数

```python
# dataframe如何访问行数据
# df1.head(5)
# df1.iloc[0]  # 返回第一行的数据
# df1.iloc[1:3]  # 返回1、2行的数据
# df1.loc['10']  # 有问题

# dataframe如何访问列数据，及特定列数据
# df1['Survived']
df1[['Survived', 'Pclass']]
```

```python
# print(np.corrcoef(df[['Survived', 'Pclass']]))   # 用的有问题
```

```python
def correlation(x, y):
    std_x = (x - x.mean()) / x.std(ddof=0)
    std_y = (y - y.mean()) / y.std(ddof=0)
    return (std_x * std_y).mean()
```

#### 计算生存与等级、年龄、性别、朋友、亲属、费用之间的皮尔逊矩相关系数

```python
print(correlation(df['Survived'], df['Pclass']))
print(correlation(df['Survived'], df['Age']))
print(correlation(df['Survived'], df['sex_index']))
print(correlation(df['Survived'], df['SibSp']))
print(correlation(df['Survived'], df['Parch']))
print(correlation(df['Survived'], df['Fare']))
```

#### 可以看到，生存率与性别的相关性比较高，且是负相关，上边对性别进行了预处理，可以看到生存率与男这个性别是负相关，由此可推断女性的生存率较高。
